# Finding Insights

## Milestone 1

```python
# Identifying countries which have low average rank but, have some very high ranking universities.

# Creating a new data set "df_countries "with columns namely - Country, Avg Rank, no. of universities in top 100.

# PHASE I
df_avg_rank = df.groupby('Country')['Global Rank'].mean().reset_index()   # I don't understand the working of this code
df_avg_rank.rename(columns={'Global Rank' : 'Avg Rank'}, inplace=True)    # I don't know where to use which column

# PHASE II
df_top_500 = df[df['Global Rank'] <= 500]
df_top_universities = df_top_500.groupby('Country')['University'].count().reset_index()
df_top_universities = df_top_universities.rename(columns={'University' : 'no. of top U'})

# PHASE III (merging)
df_countries = df_avg_rank.merge(df_top_universities, on='Country', how = 'left')

# Filling the missing values
df_countries['no. of top U'] = df_countries['no. of top U'].fillna(0)

# Sorting
df_countries = df_countries.sort_values(by='no. of top U', ascending=False)

print(df_countries.head(40))
```

```python
# Finding the University with having a value >= 0 in the column "no. of Top U" and having the worst Avg Rank
df_not_0 = df_countries[df_countries["no. of top U"] > 0]

# Now filtring down countries with worst rank
df_worst_R = df_not_0.sort_values(by='Avg Rank', ascending = False)

print(df_worst_R.head(40))
```

```python
print(df_countries.query("Country == 'India'"))
```

```python
# Creating Prime List
prime_list_1 = df_worst_R.head(20)

print(prime_list_1)
```

## Milestone 2

## Milestone 4

```python
# Defining the set of recovering countries
recovering_countries = {'Ukraine', 'Iraq', 'Sierra Leone', 'Rwanda', 'Bosnia and Herzegovina'}

# Filtering universities in recovering countries with Global Rank < 2000
recovering_df = df[(df['Country'].isin(recovering_countries)) & (df['Global Rank'] < 5000)]

# Displaying the result
print(recovering_df)
```

## Final

```python
# Creating final_list
final_list = {'Cost-Rica', 'Indonesia', 'China', 'Thailand', 'Russia', 'Ukraine', 'Rwanda'}

# Filtering the dataset for only the countries in final_list
filtered_df = df[df['Country'].isin(final_list)]

# Sorting by 'Country' and 'Global Rank' (ascending order)
filtered_df = filtered_df.sort_values(by=['Country', 'Global Rank'], ascending=[True, True])

# Getting the top 5 universities in each country
top_5_per_country = filtered_df.groupby('Country').head(5)

# Display the result
print(top_5_per_country)
```
